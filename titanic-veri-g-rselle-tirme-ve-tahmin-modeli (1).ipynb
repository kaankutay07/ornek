```python
# erksik verilerin tamamlanması
#eksik bir sayıyı tamamlamak için ortalama, ortanca, sabitr değer gibi değerler kul.
#kategorik değerlerde eksik varsa : sabit değer verilir, Frekansı en yüksek olan değer verilebilir
#eksik veriler çok sıkıntılıysa ve tamamlamaka zorsa silinebilir
```

```python
df.isnull().sum()
```

```python
#Ortanca, ortalama bilet ücreeti
df['Fare'].median(),df['Fare'].mean()
```

```python
#ortanca değeri kulalnarak eksik veirlei tamamlamak
df['Fare'].fillna(df['Fare'].median(),inplace=True)
```

```python
#Eksik veri kontrolü
df.Fare.isnull().sum()
```

```python
#KAtegorik değerlerde en çok tekraqr eden değer
df['Embarked'].mode()[0]
```

```python
df.Embarked.fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.Embarked.isnull().sum()
```

```python
#Özellik dönüştürme-özellik mühendisiliği
df['Name']
```

```python
df['Title']=df['Name'].str.extract('([A-Za-z]+)\.')
```

```python
df.Title.unique()
```

```python
df['Title']=df['Title'].replace(['Ms','Mlle'],'Miss')
df['Title']=df['Title'].replace(['Mme','Countess','Lady','Dona'],'Mrs')
df['Title']=df['Title'].replace(['Capt','Col','Don','Dr','Major','Rev','Sir','Jonkheer'],'Mr')
```

```python
df['Title'].unique()
```

```python
#yaş sütunundaki eksikler
df.Age.isnull().sum()
```

```python
df['Age'].groupby(df['Title']).mean()
```

```python
df.groupby('Title')['Age'].mean()
```

```python
#Title grup ortalmaalrına göre eksik veirlei doldur
df['Age']=df['Age'].fillna(df.groupby('Title')['Age'].transform('mean'))
```

```python
df.Age.isnull().sum()
```

```python
#toplu oalrka eksik veri kontrolü 
df.isnull().sum()
```

```python
#Feature Engineering 
#Özellik Mühendisliği-özellik çıkarımı
#Kullanıcılak özelikleri belirlemek ve gereksiz sütunları çıakrmak
df_mod=df.drop(['PassengerId','Name','Ticket','Cabin'], axis=1)
```

```python
df_mod
```

```python
# değişkenler-hedef, özleilkler-sonuç
y=df_mod['Survived']
X=df_mod.drop('Survived',axis=1)
```

```python
y
```

```python
#KAtegorik değişkenleri sayısal dseğerlere çevirdik
X=pd.get_dummies(X,drop_first=True)
X
```

```python
X.isnull().sum()
```

```python
#Eğitim ve test seti olarak ayır
X_train=X[:891]
X_test=X[891:]
y_train=y[:891]
y_test=y[891:]
```

```python
X_train,X_test
```

```python
y_train,y_test
```

```python
#veri setini bölşmek için kullanılan fonksiyon
from sklearn.model_selection import train_test_split
#modelin değerlendirilmeis için kullanılan ölçütler
from sklearn.metrics import accuracy_score, classification_report
```

```python
X_train.shape, y_train.shape
```

```python
X_train,X_val, y_train,y_val=train_test_split(X_train,y_train,
                                              test_size=.2,random_state=42)
```

```python
X_train.shape, y_train.shape, X_val.shape, y_val.shape
```

## 4. Model Oluşturma-Eğitme;

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
#modeli eğitme fit
model.fit(X_train,y_train)
```

## 5. Modeli Değerlendirme Tahmin ve Test

```python
#model tahminleri yapsın
y_pred=model.predict(X_val)
```

```python
y_pred
```

```python
accuracy_score(y_pred,y_val)
```

```python
y_test_tahmin=model.predict(X_test)
y_test_tahmin
```

```python
len(y_test_tahmin)
```
